```python
def f(n):
    i = 0
    while i < n:
        yield i
        i += 1
```

```python
gen = f(2)
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
gen = (i for i in range(10))
gen
```

```python
def cache(f):
    dct = dict()
    def ff(*args):
        if args in dct:
            return dct[args]
        else:
            dct[args] = f(*args)
            return dct[args]
    return ff

@cache
def f(x):
    return x

@cache
def g(x):
    return x*2
f(3)
f(5)
f(3)
g(3)
```
