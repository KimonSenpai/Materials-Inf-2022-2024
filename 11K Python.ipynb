```python
def my_range(n):
    i = 0
    while i < n:
        yield i
        i += 1
        
```

```python
gen = my_range(3)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def cache(f):
    dct = {}
    def ff(*args):
        if args in dct:
            return dct[args]
        else:
            dct[args] = f(*args)
            return dct[args]
    
    return ff
```

```python
@cache
def f(x):
    print("f")
    return x**2
#f = cache(f)

f(2)
f(3)
f(2)
```
